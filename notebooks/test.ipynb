````python
from langchain_core.prompts import PromptTemplate #pip install langchain_core
from langchain.output_parsers import ResponseSchema, StructuredOutputParser #pip install langchain
from langchain_openai import ChatOpenAI #pip install langchain_openai
import dotenv #pip install dotenv


dotenv.load_dotenv()

PROMPT = """
# TODO:

# REVIEW
```
{review}
```

Provide an analysis of the following review using this format instruction
{format_instructions}

"""


parser = StructuredOutputParser(
    response_schemas=[
        ResponseSchema(name="Thougths & reasoning", description="Your thoughts & reasoning here."),
        ResponseSchema(name="emotional response", description="The best-fitting emotional response this review describes."),
        ResponseSchema(name="Reason for review", description="Reasoning behind the bad or good review"),
        ResponseSchema(name="Solution for review", description="Give a possible solution for if the review has a negative feeling, else just return \"no solution needed\"")
    ]
)

template = PromptTemplate.from_template(
    PROMPT, partial_variables={"format_instructions": parser.get_format_instructions()}
)

llm = ChatOpenAI(
    model="gpt-4o-mini"
)
chain = template | llm.with_retry() | parser
````

````python
import os
os.getenv("OPENAI_API_KEY")
````

````python
chain.invoke({
    "review": 
    {   "review": "I ate the Iphone",
        "title": "My wife got unalived by this phone",
        "stars": 0
    }
})
````
